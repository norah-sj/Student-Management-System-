```python
import os

# --- 1. Fatima Abdullah (Parent Class & Encapsulation) ---
class Person:
    def __init__(self, name, person_id):
        self.name = name
        self.__id = person_id  # private attribute

    def get_id(self):
        return self.__id


# --- 2. Nora Saad (Inheritance) ---
class Student(Person):
    def __init__(self, name, person_id, grade):
        super().__init__(name, person_id)
        self.grade = grade

    def display(self):
        print("Name:", self.name, "| ID:", self.get_id(), "| Grade:", self.grade)

    def __str__(self):
        return f"{self.name},{self.get_id()},{self.grade}"


# --- 3. Raghad & Juana (File Handling) ---
def load_students():
    students = []

    if os.path.exists("students_data.txt"):
        file = open("students_data.txt", "r")

        for line in file:
            data = line.strip().split(",")

            if len(data) == 3:
                name = data[0]
                person_id = data[1]
                grade = int(data[2])

                students.append(Student(name, person_id, grade))

        file.close()

    return students


def save_students(students):
    file = open("students_data.txt", "w")

    for s in students:
        file.write(s.name + "," + s.get_id() + "," + str(s.grade) + "\n")

    file.close()


# --- 4. Nasayim & Amjad (Main Program) ---
def main():
    students_list = load_students()

    while True:
        print("\n--- Student System ---")
        print("1- Add Student")
        print("2- Show Students")
        print("3- Update Student")
        print("4- Delete Student")
        print("5- Save & Exit")

        choice = input("Choose: ")

        # ADD
        if choice == "1":
            name = input("Name: ")
            p_id = input("ID: ")
            grade = input("Grade: ")

            if not grade.isdigit():
                print("Grade must be number")
                continue

            students_list.append(Student(name, p_id, int(grade)))
            print("Added")

        # SHOW
        elif choice == "2":
            if len(students_list) == 0:
                print("No students")
            else:
                for s in students_list:
                    s.display()

        # UPDATE
        elif choice == "3":
            search_id = input("Enter ID: ")
            found = False

            for s in students_list:
                if s.get_id() == search_id:
                    s.name = input("New name: ")
                    new_grade = input("New grade: ")

                    if new_grade.isdigit():
                        s.grade = int(new_grade)

                    print("Updated")
                    found = True
                    break

            if not found:
                print("Not found")

        # DELETE
        elif choice == "4":
            search_id = input("Enter ID: ")

            new_list = []
            found = False

            for s in students_list:
                if s.get_id() == search_id:
                    found = True
                else:
                    new_list.append(s)

            students_list = new_list

            if found:
                print("Deleted")
            else:
                print("Not found")

        # SAVE & EXIT
        elif choice == "5":
            save_students(students_list)
            print("Saved, bye!")
            break

        else:
            print("Wrong choice")


if __name__ == "__main__":
    main()
```
